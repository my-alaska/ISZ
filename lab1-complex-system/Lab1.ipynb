Below are **five popular predator-prey models** commonly found in scientific research:

### 1. **Lotka-Volterra Model**
   The **Lotka-Volterra** model is the most basic and well-known predator-prey model, named after Alfred Lotka and Vito Volterra. It consists of a pair of first-order, non-linear differential equations. The model assumes that the predator and prey populations change continuously over time and interact with each other in the following way:

   - **Prey Population Dynamics**: The prey grows exponentially in the absence of predators but decreases due to predation.
   - **Predator Population Dynamics**: The predator population depends on the availability of prey and declines if prey is scarce.

   **Equations**:
   $$
   \frac{dx}{dt} = \alpha x - \beta x y
   $$
   $$
   \frac{dy}{dt} = \delta x y - \gamma y
   $$
   where:
   - $x $ = prey population
   - $y $ = predator population
   - $\alpha $ = prey growth rate
   - $\beta $ = predation rate
   - $\delta $ = predator reproduction rate (dependent on prey)
   - $\gamma $ = predator mortality rate

   **Key Features**: Cyclic oscillations of predator and prey populations, which never stabilize without additional factors like carrying capacity.

---

### 2. **Rosenzweig-MacArthur Model**
The **Rosenzweig-MacArthur** model extends the basic Lotka-Volterra model by incorporating a **saturating functional response** for predators, which means that predation rates increase more slowly as prey density increases. This addresses one of the shortcomings of the Lotka-Volterra model by limiting the rate at which predators consume prey when prey is abundant, preventing runaway predation.

   **Equations**:
   $$
   \frac{dx}{dt} = r x \left(1 - \frac{x}{K}\right) - \frac{p x}{h + x}
   $$

   $$
   \frac{dy}{dt} = s y \left(\frac{p x}{h + x}\right) - d y
   $$

   where:
   - $ r $ = intrinsic growth rate of prey
   - $ K $ = carrying capacity of the environment
   - $ p $ = maximum predation rate
   - $ h $ = prey density at which predation is half-maximal
   - $ s $ = efficiency with which prey is converted to predator offspring
   - $ d $ = predator death rate

   **Key Features**: The introduction of carrying capacity for prey and predator saturation leads to more realistic dynamics, such as stable equilibria or complex cycles like the "paradox of enrichment," where increasing prey availability can destabilize the system.

---

### 3. **Holling's Type II Functional Response Model**
   Developed by **C.S. Holling**, this model emphasizes the **functional response** of predators, particularly how the rate of prey consumption by predators slows down as prey density increases. This is called **Type II functional response** and addresses how handling time (time required to catch and consume prey) limits predation.

   **Equation**:
   $$
   \frac{dy}{dt} = \frac{a x}{1 + b x}
   $$

   where:
   - $ a $ = maximum predation rate
   - $ b $ = prey population at which predation rate is half its maximum
   - $ x $ = prey population
   - $ y $ = predator population

   **Key Features**: This model explains why predators might not increase their prey consumption linearly even when prey numbers rise, introducing predator saturation at high prey densities.

---

### 4. **Beddington-DeAngelis Model**
   The **Beddington-DeAngelis** model modifies the functional response in predator-prey interactions by incorporating a **mutual interference** term among predators, meaning that predator efficiency decreases when there are too many predators competing for the same prey. It also considers crowding effects among prey populations.

   **Equations**:
   $$
   \frac{dx}{dt} = r x \left(1 - \frac{x}{K}\right) - \frac{p x}{h + x + ay}
   $$

   $$
   \frac{dy}{dt} = s y \left(\frac{p x}{h + x + ay}\right) - d y
   $$

   where:
   - $ a $ = interference between predators.

   **Key Features**: This model accounts for the **negative feedback** effects when predators interfere with one another, providing a more nuanced view of real-world predator-prey systems where predators are not perfectly efficient at high densities.

---

### 5. **Nicholson-Bailey Model**
   The **Nicholson-Bailey** model is a **discrete-time model** often used to describe predator-prey dynamics in systems with distinct generations (e.g., insects with seasonal reproduction). The model is particularly suitable for host-parasite or prey-predator systems with time lags.

   **Equations**:
   $$
   N_{t+1} = N_t e^{r(1 - N_t / K)} \cdot e^{-aP_t}
   $$

   $$
   P_{t+1} = c N_t (1 - e^{-aP_t})
   $$

   where:
   - $ N_t $ = prey population at time $ t $
   - $ P_t $ = predator population at time $ t $
   - $ r $ = prey growth rate
   - $ K $ = carrying capacity
   - $ a $ = attack rate of predators
   - $ c $ = conversion efficiency of prey into predator offspring.

   **Key Features**: This model produces oscillations and can exhibit chaotic behavior. It is particularly suited to studying periodic fluctuations in prey and predator populations due to seasonal reproduction or other time-lagged processes.

---

### Summary:
These five predator-prey models each provide a different perspective on the interaction between predator and prey populations, from basic continuous-time models (e.g., Lotka-Volterra) to more realistic descriptions (e.g., Rosenzweig-MacArthur) and discrete-time models (e.g., Nicholson-Bailey). Each model has its advantages, depending on the ecological scenario being studied, and they are valuable tools in the analysis of complex systems in ecology.

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parameters of the system
alpha = 1.0  # Prey growth rate
beta = 0.1  # Rate of predation
delta = 0.075  # Predator reproduction rate
gamma = 1.5  # Predator death rate
epsilon = 0.01  # Rate of apex predator consuming intermediate predator
eta = 0.02  # Rate of apex predator reproduction
zeta = 0.5  # Apex predator death rate


# Lotka-Volterra system with 3 dynamic variables
def lotka_volterra(t, vars):
    x, y, z = vars
    dxdt = alpha * x - beta * x * y
    dydt = delta * x * y - gamma * y - epsilon * y * z
    dzdt = eta * y * z - zeta * z
    return [dxdt, dydt, dzdt]


# Initial conditions (population of prey, intermediate predator, and apex predator)
x0 = 40  # Initial prey population
y0 = 9  # Initial intermediate predator population
z0 = 5  # Initial apex predator population
initial_conditions = [x0, y0, z0]

# Time points where the solution is computed
t_span = (0, 100)  # Simulate from t=0 to t=100
t_eval = np.linspace(0, 100, 1000)  # 1000 time points for smooth plots

# Solve the differential equations
solution = solve_ivp(lotka_volterra, t_span, initial_conditions, t_eval=t_eval)

# Extract solutions
x_vals = solution.y[0]  # Prey population
y_vals = solution.y[1]  # Intermediate predator population
z_vals = solution.y[2]  # Apex predator population

# Plot the results in a 3D phase space
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection="3d")

# Plotting (x, y, z) - Prey, Intermediate Predator, Apex Predator in phase space
ax.plot(x_vals, y_vals, z_vals, lw=1.5, color="b")

# Labels and title
ax.set_xlabel("Prey (x)")
ax.set_ylabel("Intermediate Predator (y)")
ax.set_zlabel("Apex Predator (z)")
ax.set_title("Lotka-Volterra System with 3 Dynamic Variables")

plt.show()
```
