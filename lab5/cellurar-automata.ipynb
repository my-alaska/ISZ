```python
import cellpylib as cpl
import numpy as np

simple = np.zeros(128)[None,:]
simple[:,64] = 1

random = np.zeros(128)[None,:]
random[:,np.random.randint(0,128,128)] = 1

for rule in [30,45,106,164]:
    for start_state in [simple,random]:
    
        cellular_automaton = cpl.evolve(
            start_state,
            timesteps=128,
            memoize=True,
            apply_rule=lambda n, c, t: cpl.nks_rule(n, rule),
        )
        
        cpl.plot(cellular_automaton)
```
